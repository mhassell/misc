```python
# let's try pulling beej's guide page by page
from lxml import html
import requests
```

```python
page = requests.get('http://beej.us/guide/bgnet/output/html/multipage/index.html')
tree = html.fromstring(page.content)
```

```python
#This will create a list of td's:
tds = tree.xpath('//td')
startpath = 'http://beej.us/guide/bgnet/output/html/multipage/index.html'
currentPath = startpath
mainpath = 'http://beej.us/guide/bgnet/output/html/multipage/'
endpath = 'http://beej.us/guide/bgnet/output/html/multipage/zindex.html'
paths = []
paths.append(startpath)
```

```python
while currentPath != endpath:
    page = requests.get(currentPath)
    tree = html.fromstring(page.content)
    tds = tree.xpath('//td')
    for i in tds:
        if i.values()[0]=='right':
            currentPath = mainpath+i[0].values()[0]
            paths.append(currentPath)
            break
```

```python
options = {
    'page-size': 'Letter',
    'margin-top': '0.75in',
    'margin-right': '0.75in',
    'margin-bottom': '0.75in',
    'margin-left': '0.75in',
    'javascript-delay' : 2000,
    'minimum-font-size': 512  # why the hell is the font so small?
}
pdfkit.from_url(paths, 'beejNetwork2.pdf', options=options)
```
