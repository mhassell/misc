```python
# this was too easy...need to learn something from this whole experiment
import pdfkit
options = {
    'page-size': 'Letter',
    'margin-top': '0.75in',
    'margin-right': '0.75in',
    'margin-bottom': '0.75in',
    'margin-left': '0.75in',
    'minimum-font-size': 512  # why the hell is the font so small?
}
pdfkit.from_url('http://beej.us/guide/bgnet/output/html/singlepage/bgnet.html', 'beejNetwork.pdf',options=options)
```

```python
# let's try pulling beej's guide page by page
from lxml import html
import requests
```

```python
page = requests.get('http://beej.us/guide/bgnet/output/html/multipage/index.html')
tree = html.fromstring(page.content)
```

```python
#This will create a list of buyers:
buyers = tree.xpath('//div[@title="buyer-name"]/text()')
#This will create a list of prices
prices = tree.xpath('//span[@class="item-price"]/text()')
```
